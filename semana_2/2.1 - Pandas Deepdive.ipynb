### AplicaciÃ³n de funciones

````python
df.head()
````

````python
def suma(x):
    return x+2
````

````python
%%time

df.quantity = [suma(x) for x in df.quantity]

df.head()
````

````python
%%time

df.quantity = list(map(suma, df.quantity))

df.head()
````

````python
%%time

df.quantity = pd.Series(list(map(lambda x: x+2, df.quantity)))

df.head()
````

````python
%%time

df.quantity = df.quantity.apply(suma)  # esto es un map, el iterable es la columna

df.head()
````

````python
%%time

df.quantity = df.quantity.apply(lambda x: x+2)  # esto es un map, el iterable es la columna

df.head()
````

````python
%%time

df.quantity = df.quantity + 2

df.head()
````

````python
%%time

df['nueva'] = df.quantity / df.unitprice

df.head()
````
